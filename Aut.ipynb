```
import networkx as nx
import itertools
```

# An intuitive solution but brute force

### Instead of generating all possible permutations of the graph and determining if each one is an isomorphism with the original, what you do is generate all permutations that retain the shape of the graph.


## That is: Permutations are generated only on nodes of the same degree, since they are the only ones that do not alter the original shape when permuted!






#Input graph

```
nodes = input('Ingrese los nodos del grafo separados por comas: ')
nodes = nodes.split(',')
nodes = [int(n) for n in nodes]

edges_count = input('Ingrese el número de aristas del grafo: ')
while not edges_count.isdigit():
    edges_count = input('El número que ingresó no es un número válido. Ingrese de nuevo el número de aristas del grafo: ')
edges_count = int(edges_count)

edges = []
for i in range(0,edges_count):
    edge = input('Ingrese el par de vértices para la arista separados por comas (#{}): '.format(i+1))
    edge = edge.split(',')
    edges.append((int(edge[0]), int(edge[1])))
```

# Set graph

```
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

# Edges count by node

```
m = []
for i in nodes:
    m.append(len(G.edges(i)))
```

#Same grade nodes

```
nodes_same_grade = [[] for i in range(max(m)+1)]
for i in nodes:
    temp = nodes_same_grade[len(G.edges(i))].append(i)
nodes_same_grade = [n for n in nodes_same_grade if n!=[]]
```

```
def combinePermutations(perms):
    a = min(min(p) for p in perms)
    b = max(max(p) for p in perms)
    d = {}
    for p in perms:
        pairs = zip(sorted(p),p)
        for i,j in pairs:
            d[i] = j
    return tuple(d[i] for i in range(a,b+1))
```

```
def listPerms(cells):
    products = itertools.product(*(itertools.permutations(cell) for cell in cells))
    return [combinePermutations(perms) for perms in products]
```

# Result:


## The result is a list of possible permutations of the graph that retain the original shape of the graph, I mean the group of automorphisms in the graph.


```
for p in listPerms(nodes_same_grade):
  print(p)
```
